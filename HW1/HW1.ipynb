```python
import numpy as np
import pandas as pd
```

```python
data = np.array([1,2,3,4,5])
print(f'mean = {data.mean()}')
print(f'variance = {data.var(ddof=0)}')
```

```python
# https://stackoverflow.com/questions/1208118/using-numpy-to-build-an-array-of-all-combinations-of-two-arrays
samples = np.array(np.meshgrid(data, data)).T.reshape(-1,2)
df = pd.DataFrame(samples, index=range(1,len(samples)+1), columns=['num1', 'num2'])
K = len(df)
print(f'{K = }')
df
```

```python
df['mean'] = df[['num1', 'num2']].mean(axis=1)
df['variance'] = df[['num1', 'num2']].var(axis=1, ddof=1)
df
```

```python
x_dot = df['mean'].mean()  # 1
var_bar_1 = df['variance'].sum() / K  # 2
df['msd'] = df.apply(lambda row : (row['num1'] - row['num2'])**2/2, axis=1)
var_bar_2 = df['msd'].sum() / K  # 3
print(f'{x_dot = }')
print(f'{var_bar_1 = }')
print(f'{var_bar_2 = }')

df
```

```python
all(df.variance == df.msd)
```
